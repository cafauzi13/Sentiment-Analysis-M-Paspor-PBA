#### 2. Betahita: https://betahita.id/news/detail/10512/transisi-energi-dihambat-bank-nasional-riset.html
#### 3. Tempo: https://www.tempo.co/ekonomi/pegiat-环境-market-forces-sebut-ada-5-bank-danai-proyek-smelter-dan-pltu-adaro-169307

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

```

```python
# 1. Dataset dari Artikel Berita (CNBC, Betahita, Tempo)
corpus = [
    "pungut salur iuran batu bara bakal dikelola oleh bank ri", # CNBC
    "transisi energi dihambat bank nasional riset pembiayaan batu bara", # Betahita
    "pegiat lingkungan sebut bank danai proyek smelter pltu adaro batu bara" # Tempo
]

# 2. Representasi Vektor (Fokus pada kata kunci yang sama di 3 artikel: bank & batu bara)
vectorizer = CountVectorizer(vocabulary=['bank', 'batu', 'bara'])
X = vectorizer.fit_transform(corpus)

```

```python
# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen dengan Dokumen 0 (CNBC)
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['bank', 'batu', 'bara'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

```python
import pandas as pd
from collections import Counter
import numpy as np

# Data Artikel
documents = [
    "pungut salur iuran batu bara bakal dikelola oleh bank ri",
    "transisi energi dihambat bank nasional riset pembiayaan batu bara",
    "pegiat lingkungan sebut bank danai proyek smelter pltu adaro batu bara"
]

# Membuat DataFrame TF secara manual (Persis template)
tf = pd.DataFrame(
    [Counter(doc.split()) for doc in documents],
).fillna(0)

# Fungsi hitung manual (Persis template)
def length(v):
  return np.sqrt((v ** 2).sum())

def cos_dist(v, w):
  return 1 - (v * w).sum() / (length(v) * length(w))

# Hasil perbandingan (CNBC ke Betahita, CNBC ke Tempo)
print("Jarak Cosine (Manual):")
print(cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2]))
```

**CountVectorizer:** Mengubah teks menjadi angka (matriks frekuensi kata). Di sini kita membatasi fitur hanya pada kata "hate" dan "darkness" agar sesuai dengan grafik di slide Stanford.

**Cosine Similarity vs Distance:** Fungsi ini menghasilkan nilai 0 sampai 1. Semakin mendekati 1, semakin mirip kedua dokumen tersebut.

**Analisis Hasil:** Bahwa Dokumen 2 (MLK) memiliki nilai similarity yang tinggi terhadap Dokumen 0, membuktikan bahwa meskipun jumlah katanya jauh lebih banyak, "arah" topiknya tetap serupa.

**Euclidean Distance:** Jika kita menggunakan jarak lurus, Dokumen 2 akan dianggap sangat berbeda karena jumlah kemunculan katanya (3 dan 4) jauh dari Dokumen 0 (1 dan 2).

**Aplikasi di Dunia Nyata:** Teknik ini adalah dasar dari mesin pencari (search engines) dan sistem rekomendasi teks yang sering kita gunakan sehari-hari.

```python
import pandas as pd
from collections import Counter

# 1. Gunakan data ulasan artikel
documents_news = [
    "pungut salur iuran batu bara bakal dikelola oleh bank ri", # Doc 0 (CNBC)
    "transisi energi dihambat bank nasional riset pembiayaan batu bara", # Doc 1 (Betahita)
    "pegiat lingkungan sebut bank danai proyek smelter pltu adaro batu bara" # Doc 2 (Tempo)
]

# 2. Membuat Matriks Frekuensi (TF) secara manual untuk artikel
# Ini akan menghitung SEMUA kata unik yang ada di artikel tersebut
tf_news = pd.DataFrame(
    [Counter(doc.split()) for doc in documents_news],
).fillna(0).astype(int)

print("--- Matriks Frekuensi Kata (TF) Manual Artikel ---")
display(tf_news)

# 3. Hitung Jarak Cosine Manual (Menggunakan fungsi cos_dist yang sudah dibuat di atas)
# Membandingkan CNBC dengan Betahita dan CNBC dengan Tempo
print("\n--- Jarak Cosine Manual Antar Artikel ---")
print(f"Jarak CNBC ke Betahita: {cos_dist(tf_news.loc[0], tf_news.loc[1])}")
print(f"Jarak CNBC ke Tempo: {cos_dist(tf_news.loc[0], tf_news.loc[2])}")
```
