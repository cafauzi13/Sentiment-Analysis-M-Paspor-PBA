## WEEK 3: Bag of Words (BoW) - News Article Analysis

### SUMBER ARTIKEL:

1. CNBC: https://www.cnbcindonesia.com/news/20231121154850-4-490788/pungut-salur-iuran-batu-bara-bakal-dikelola-oleh-3-bank-ri
2. Betahita: https://betahita.id/news/detail/10512/transisi-energi-dihambat-bank-nasional-riset.html
3. Tempo: https://www.tempo.co/ekonomi/pegiat-环境-market-forces-sebut-ada-5-bank-danai-proyek-smelter-dan-pltu-adaro-169307

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from collections import Counter
import re
from sklearn.feature_extraction.text import CountVectorizer

# Download dataset pendukung
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Step 1: Tokenize data, remove stop words and perform stemming/lemmatization.
# Isi paragraph diambil dari artikel CNBC Indonesia yang Nisa berikan
paragraph = """
Pemerintah akan memungut dan menyalurkan iuran batu bara melalui skema Mitra Instansi Pengelola (MIP).
Tiga bank besar yakni Bank Mandiri, BNI, dan BRI bakal dilibatkan untuk mengelola dana kompensasi tersebut.
Skema ini bertujuan untuk menjamin ketersediaan pasokan batu bara di dalam negeri terutama untuk kebutuhan PLN.
"""

sentences = nltk.sent_tokenize(paragraph)
lemmatizer = WordNetLemmatizer()
corpus = []

# Proses pembersihan teks sesuai template
for i in range(len(sentences)):
    # Membersihkan karakter non-huruf
    sent = re.sub('[^a-zA-Z]', ' ', sentences[i])
    sent = sent.lower()
    sent = sent.split()
    # Menghapus stopword (menggunakan English sesuai template, atau ganti 'indonesian')
    sent = [lemmatizer.lemmatize(word) for word in sent if not word in set(stopwords.words('indonesian'))]
    sent = ' '.join(sent)
    corpus.append(sent)

print("--- Step 1: Corpus ---")
print(corpus)
```

```python
# Step 2: List all unique words
all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

unique_words = list(set(all_words))
print("\n--- Step 2: Unique words ---")
print(unique_words)
```

```python
# Step 3: Create a dictionary mapping of words to a number (Sorted by Frequency)
word_frequencies = Counter(all_words)

print("\n--- Step 3: Word Frequencies ---")
print(word_frequencies)

# Sorting words by frequency (descending)
sorted_words = sorted(word_frequencies.items(), key=lambda item: item[1], reverse=True)

# Creating the dictionary mapping words to numbers
word_to_number = {}
for i, (word, frequency) in enumerate(sorted_words):
    word_to_number[word] = i + 1

print("\nWord to Number Mapping (Sorted by Frequency):")
print(word_to_number)

# Ekstraksi Fitur menggunakan Scikit-Learn (Sesuai Template)
cv = CountVectorizer()
independentFeatures = cv.fit_transform(corpus).toarray()

print("\n--- Independent Features (BoW Matrix) ---")
print(independentFeatures)
```

```python
"""
Teks merupakan paragraf yang menceritakan tentang pengelolaan iuran bank.
Jika kita mencari term bank, maka akan muncul juga pencarian bank-bank (plural).
"""

# 1. Mencari kata "bank" (mirip contoh "towel" di template)
print("Hasil grep 'bank':")
!grep "bank" sample_data/artikel_berita.txt

# 2. Mencari variasi penulisan "Batu Bara" (Hitchhiker diganti Batu Bara)
# Pola ini mencari: Batu Bara, batubara, batu-bara, dll.
print("\nHasil grep 'Batu Bara':")
!grep "\(B\|b\)atu *\(B\|b\)ara" sample_data/artikel_berita.txt

# --- IMPLEMENTASI DENGAN LIBRARY RE (PYTHON) ---

# Membaca file artikel sebagai string
with open('sample_data/artikel_berita.txt') as f:
    news_string = f.read()

# 1. Define the regex pattern (Mencari baris mengandung Bank)
pattern = '.*[B|b]ank.*'
regex = re.compile(pattern)

print("\nRegex Findall (Baris mengandung Bank):")
print(regex.findall(news_string))

# 2. Define the regex pattern (Mencari variasi Batu Bara dengan Grouping)
# Pola: Mencari kata Batu/batu yang diikuti spasi/strip (opsional) lalu Bara/bara
pattern_group = '(.*(([B|b]atu[- ]*[B|b]ara)|batubara).*)'
regex_group = re.compile(pattern_group)

print("\nRegex Findall (Variasi Batu Bara dengan Grouping):")
# Hasilnya akan menangkap 'batu bara', 'batubara', dan 'batu-bara'
print(regex_group.findall(news_string))
```

## WEEK 4: TF-IDF - News Article Analysis

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Inisialisasi TfidfVectorizer
# smooth_idf=False dan norm=None digunakan agar hitungannya murni/manual
vec_tfidf = TfidfVectorizer(smooth_idf=False, norm=None)

# 2. Fit dan Transform menggunakan korpus artikel berita (dari Week 3)
# Pastikan variabel 'corpus' sudah terdefinisi di bagian BoW sebelumnya
tfidf_matrix = vec_tfidf.fit_transform(corpus)

# 3. Tampilkan hasil TF-IDF dalam bentuk matriks dense/array
print("Matriks TF-IDF (Raw Values):")
print(tfidf_matrix.toarray())

# 4. Merubah ke DataFrame agar rapi
# Ini memudahkan kita melihat bobot tiap kata di setiap dokumen
df_tfidf_news = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vec_tfidf.get_feature_names_out(),
    index=['Artikel 1 (CNBC)', 'Artikel 2 (Betahita)', 'Artikel 3 (Tempo)']
)

print("\n--- Tabel Bobot TF-IDF Artikel Berita ---")
display(df_tfidf_news)
```
